# 1.Deteksi Warna Pada Citra

```python
import cv2 #ini merupakan code digunakan untuk membaca, menulis, menampilkan, memproses, dan menganalisis gambar dan video.
import matplotlib.pyplot as plt #merupakan code yang digunakan untuk membuat grafik dan plot untuk memvisualisasikan data.
import numpy as np #digunakan untuk melakukan operasi matematika pada data, seperti penjumlahan, pengurangan, perkalian, dan pembagian.
```

```python
img = cv2.imread("pcdd.jpg") #disini kita akan memanggil sebuah gambah yang bernama pccd.jpg
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #ini merupakan proses untuk mengubah gambar dari BGR ke RGB
plt.imshow(rgb)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #ini adalah code yang digunakan untuk mengubah gambar dari BGR ke Gray
alpha = 1.0 #Ini merupakan code untuk membuat kontras yang dimana jika semakin tinggi angkanya maka akan semakin tinggi tinkat kontrasnya
beta = 0 #code ini berfungsi untuk menambahkan kecerahan pada gambar
adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta) #dan pada code berikut gunanya untuk mengontrol kontras dan kecerahan pada gambar
plt.figure(figsize=(10,5)) #disini saya membuat plot dengan menentukan ukuran yang saya inginkan
plt.subplot(1,2,2) #dan ini adalah subplot yang berada didalam plot
plt.title('citra kontras') #untuk memberikan judul pada gambar saat akan di cetak
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #disini saya menggunakan kode ini untuk Fungsi ini mengubah ruang warna gambar.
alpha = 1.7 #dan saya menerapakn 1.7 untuk membuat warna kontras pada gambar
beta = 1 #dan 1 untuk kecerahan gambar
adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

plt.figure(figsize=(10,5))
plt.subplot(1,2,2)
plt.title('Citra Kontras')
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.axis('off')

adjusted_rgb = cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB) #adjusted disini berfungsi untuk menyimpan salinan gambar adjusted yang sudah dalam format RGB.
merah_adjusted = adjusted_rgb[:,:,0] # Variabel ini menyimpan kanal merah dari gambar adjusted_rgb dengan . Pengindeksan [:,:,0]
hijau_adjusted = adjusted_rgb[:,:,1] # Variabel ini menyimpan kanal hijau dari gambar adjusted_rgb dengan . Pengindeksan [:,:,1]
biru_adjusted = adjusted_rgb[:,:,2] # Variabel ini menyimpan kanal biru dari gambar adjusted_rgb dengan . Pengindeksan [:,:,2]

fig, axis = plt.subplots(1, 3, figsize=(10,10))
ax = axis.ravel()

ax[0].imshow(merah_adjusted, cmap='gray') #untuk mencetak gambar warna merah
ax[0].set_title('Red Color')
ax[1].imshow(hijau_adjusted, cmap='gray') #untuk mencetak gambar warna hijau
ax[1].set_title('Green Color')
ax[2].imshow(biru_adjusted, cmap='gray') #untuk mencetak gambar warna hijau
ax[2].set_title('Blue Color')

plt.show()
```

# 2. MENGURUTKAN AMBANG BATAS TERKECIL SAMPAI DENGAN TERBESAR

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Membaca gambar RGB
rgb = cv2.imread('pcdd.jpg')

# Konversi gambar menjadi grayscale
gray_image = cv2.cvtColor(rgb, cv2.COLOR_BGR2GRAY)

# Mengubah semua nilai piksel menjadi 0 (hitam)
black_image = np.zeros(gray_image.shape, dtype=np.uint8)

# Menampilkan gambar hitam
plt.imshow(black_image, cmap='gray')
plt.title('none')
plt.xlabel('')  # Menambahkan label sumbu x
plt.ylabel('')  # Menambahkan label sumbu y
plt.axis('on')  # Menampilkan sumbu x dan y
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar
img = cv2.imread('pcdd.jpg')

# Konversi gambar ke ruang warna HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Tentukan rentang warna biru dalam HSV
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])

# Buat mask untuk warna biru
mask = cv2.inRange(hsv, lower_blue, upper_blue)

# Aplikasikan mask ke gambar asli
blue_detected_img = cv2.bitwise_and(img, img, mask=mask)

# Tampilkan gambar asli dan hasil deteksi warna biru menggunakan plot Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blue_detected_img, cv2.COLOR_BGR2RGB))
plt.title('Blue')
plt.xlabel('')  # Menambahkan label sumbu x
plt.ylabel('')  # Menambahkan label sumbu y
plt.axis('on')  # Menampilkan sumbu x dan y
plt.show()
```

```python
# Konversi gambar ke ruang warna HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Tentukan rentang warna biru dalam HSV
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])

# Buat mask untuk warna biru
blue_mask = cv2.inRange(hsv, lower_blue, upper_blue)

# Tentukan rentang warna merah dalam HSV
lower_red = np.array([0, 50, 50])
upper_red = np.array([10, 255, 255])

# Buat mask untuk warna merah (bagian yang mendekati nilai rendah dan tinggi di ruang HSV)
red_mask_low = cv2.inRange(hsv, lower_red, upper_red)

# Tentukan rentang warna merah dalam HSV (bagian yang mendekati nilai tinggi dan rendah di ruang HSV)
lower_red2 = np.array([170, 50, 50])
upper_red2 = np.array([180, 255, 255])

# Buat mask untuk warna merah (bagian yang mendekati nilai tinggi dan rendah di ruang HSV)
red_mask_high = cv2.inRange(hsv, lower_red2, upper_red2)

# Gabungkan kedua mask merah
red_mask = cv2.bitwise_or(red_mask_low, red_mask_high)

# Gabungkan mask biru dan mask merah
red_blue_mask = cv2.bitwise_or(blue_mask, red_mask)

# Aplikasikan mask ke gambar asli
red_blue_detected_img = cv2.bitwise_and(img, img, mask=red_blue_mask)

# Tampilkan gambar asli dan hasil deteksi warna merah-biru menggunakan plot Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(red_blue_detected_img, cv2.COLOR_BGR2RGB))
plt.title('Red-Blue')
plt.xlabel('')  # Menambahkan label sumbu x
plt.ylabel('')  # Menambahkan label sumbu y
plt.axis('on')  # Menampilkan sumbu x dan y

plt.show()
```

```python
# Tentukan rentang warna merah dalam RGB
lower_red = np.array([0, 0, 100])
upper_red = np.array([100, 100, 255])

# Buat mask untuk warna merah
red_mask = cv2.inRange(img, lower_red, upper_red)

# Tentukan rentang warna hijau dalam RGB
lower_green = np.array([0, 100, 0])
upper_green = np.array([100, 255, 100])

# Buat mask untuk warna hijau
green_mask = cv2.inRange(img, lower_green, upper_green)

# Tentukan rentang warna biru dalam RGB
lower_blue = np.array([100, 0, 0])
upper_blue = np.array([255, 100, 100])

# Buat mask untuk warna biru
blue_mask = cv2.inRange(img, lower_blue, upper_blue)

# Gabungkan ketiga mask
rgb_mask = cv2.bitwise_or(red_mask, cv2.bitwise_or(green_mask, blue_mask))

# Aplikasikan mask ke gambar asli
rgb_detected_img = cv2.bitwise_and(img, img, mask=rgb_mask)

# Tampilkan gambar asli dan hasil deteksi warna merah-hijau-biru menggunakan plot Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rgb_detected_img, cv2.COLOR_BGR2RGB))
plt.title('Red-Green-Blue')
plt.xlabel('')  # Menambahkan label sumbu x
plt.ylabel('')  # Menambahkan label sumbu y
plt.axis('on')  # Menampilkan sumbu x dan y

plt.show()
```
